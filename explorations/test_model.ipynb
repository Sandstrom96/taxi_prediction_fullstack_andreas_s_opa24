```python
import joblib

model = joblib.load("../src/taxipred/data/model.joblib")
scaler = joblib.load("../src/taxipred/data/scaler.joblib")
encoded_columns = joblib.load("../src/taxipred/data/encoded_columns.joblib")

model
```

```python
encoded_columns
```

```python
import pandas as pd
new_trips = pd.DataFrame({
    "Base_Fare": [3.5],
    "Trip_Distance_km": [19],
    "Time_of_Day": ["Afternoon"],
    "Day_of_Week": ["Unknown"]
})
# Had to use drop_first=False to get the hot-encoded columns
X_new = pd.get_dummies(new_trips, columns=["Time_of_Day", "Day_of_Week"], drop_first=False)
X_new = X_new.reindex(columns=encoded_columns, fill_value=0)
print(X_new)
if scaler is not None:
    X_new = scaler.transform(X_new)
prediction = model.predict(X_new)
```
